```python
import numpy as np
import time
```

numpy生成随机数以及复制的一些细节

```python
ar1 = np.random.randint(10,20,(2,3))#产生10-20的数字，shape是2行3列
```

```python
print(ar1)
```

```python
ar2 = np.random.rand(3,4)
t = 100*np.round(ar2,5)
print(t,'............')
print(ar2)
```

```python
ar2 = np.random.rand(2,3,4)#生成0-1的浮点数（随机分布） 2维，3行，4列
print(ar2)
```

```python
ar2 = np.random.rand(2,3,4,5)#生成0-1的浮点数（随机分布） 2维，3行，4列，且生成两次
print(ar2)
```

```python
np.random.seed(20)#先设置随机种子
#之后在使用产生随机数的函数
t = np.random.randint(2,10,(2,3))
#此时会发现产生的数字一直不变，是因为给的随机数种子是一个定值，所以，一般可以使用时间来当作随机数种子
print(t)
```

```python
#实例
np.random.seed(int(time.time()))#使用时间戳作为随机数种子
t = np.random.randint(2,10,(2,3))
print(t)
```

```python
#此外还有一些生成随机数的方式
#生成均匀分布的数组
ar7 = np.random.uniform(1,10,(2,3))#.uniform(low,high,(shape))
print(ar7)
```

```python
#.normal(low,high,(shape))按照标准正态分布生成随机数
ar8 = np.random.normal(1,10,(2,3))
print(ar8)
```

拷贝的知识

```python
ar4 = ar2
ar2+=ar2
print(ar4)#潜拷贝，ar2改变。ar4会跟着改变
```

```python
ar5 = ar2[:]
ar2+=ar2
print(ar5)
ar5+=ar5
print(ar2)#ar5会复制ar2，但ar2和ar5还是会相互影响
```

```python
ar6 = ar2.copy()#此时才是创建一个新的数组
ar2+=ar2
print(ar2,'\n.........................')
print(ar6)
```
