```python
from matplotlib import pyplot as plt
```

```python
x = range(2,20,2)
```

```python
y = [1,2,3,4,5,6,7,8,9]
```

```python
fig = plt.figure(figsize=(20,8),dpi=80)
#定义图片的尺寸，dpi是像素数量，越多，当放大的时候锯齿就会越小
#图片定义必须在绘图之前
```

```python
plt.plot(x,y)
```

```python
plt.savefig("./demo2.png")
#保存图片必须在绘画之后，也是plot之后，但不能放在show之后，否则会出现一张空图
#可以保存为.svg格式，.svg是矢量图格式
```

```python
plt.show()
```
