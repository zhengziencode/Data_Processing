```python
import tensorflow as tf
```

```python
#使用张量方式创建全连接神经网络
x = tf.random.normal([2, 784])#一共两个样本，每个样本有784个特征值，shape为[b d_in],所以w1的输入节点也是784
w1 = tf.Variable(tf.random.truncated_normal([784,256], stddev=0.1))#设置输出的节点数为256 w1的shape为[d_in d_out]
b1 = tf.Variable(tf.zeros([256]))#b的shape是[d_out]也就是对应w1的shape[d_in d_out]
```

```python
o1 = tf.nn.relu(tf.matmul(x,w1)+b1)#先使用感知机获得线性模型，再使用激活函数
```

```python
o1
```

```python
#使用层方式实现
x = tf.random.normal([4, 28*28])#4个样本，28*28个特征
from tensorflow.keras import layers #导入层模块
```

```python
#layers.Dense(units,activation),units为输出的节点数,activation需指定激活函数，返回一个全连接层 实例
fc = layers.Dense(512, activation=tf.nn.relu)#512是输出节点数
```

```python
h1 = fc(x)#通过一个全连接层实例完成一次全连接层计算，返回输出张量
```

```python
h1  #输出shape为[b,d_out]
```

```python
fc.kernel#获取权重值矩阵的shape
```

```python
fc.bias#获取偏置矩阵
```

```python
fc.trainable_variables #返回待优化参数列表，此例子中就是w1和b
```
